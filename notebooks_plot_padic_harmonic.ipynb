```python
import matplotlib.pyplot as plt
import math

# Define the prime number for p-adic arithmetic
p = 7

# Define the size of the harmonic structure
harmonic_structure_size = 7

# Generate the p-adic numbers for the harmonic structure
harmonic_structure_p_adic = []
for i in range(harmonic_structure_size):
    p_adic = sum([p ** (i - j) for j in range(i + 1)])
    harmonic_structure_p_adic.append(p_adic)

# Generate the logarithmic p-adic numbers for the harmonic structure
harmonic_structure_p_adic_log = [math.log(p_adic, p) for p_adic in harmonic_structure_p_adic]

# Convert the p-adic numbers to pitch class integers
harmonic_structure = [int(p_adic % 12) for p_adic in harmonic_structure_p_adic]

# Define the custom note mapping for a hypothetical non-western scale
note_mapping = {
    0: 'C',
    1: 'C#',
    2: 'D',
    3: 'D#',
    4: 'E',
    5: 'F',
    6: 'F#',
    7: 'G',
    8: 'G#',
    9: 'A',
    10: 'A#',
    11: 'B'
}

# Function to generate note names from pitch classes using a custom mapping with modulo operation
def generate_notes(scale, mapping):
    # Create an empty list to store the note names
    notes = []
    # Loop through the pitch classes in the scale
    for pitch_class in scale:
        # Get the octave number by dividing pitch class by 12 and taking the floor
        octave = pitch_class // 12
        # Apply modulo operation to restrict pitch class within octave range
        pitch_class %= 12
        # Look up the corresponding note letter from the mapping
        note = mapping[pitch_class]
        # Concatenate the note letter with the octave number
        note_with_octave = note + str(octave)
        # Append the note with octave to the list of notes
        notes.append(note_with_octave)
    # Return the list of notes
    return notes

# Generate the notes for the harmonic scale
harmonic_scale = harmonic_structure
harmonic_notes = generate_notes(harmonic_scale, note_mapping)

# Modify the harmonic scale by adding an offset
harmonic_scale_offset = [pitch_class + 3 for pitch_class in harmonic_structure]

# Generate the notes for the harmonic scale with an offset
harmonic_scale_offset = harmonic_scale_offset
harmonic_notes_offset = generate_notes(harmonic_scale_offset, note_mapping)

# Create a scatter plot to represent p-adic numbers graphically
fig, ax = plt.subplots()
ax.scatter(harmonic_structure_p_adic_log, harmonic_structure, c='b', label='Harmonic Structure')
ax.scatter(harmonic_structure_p_adic_log, harmonic_scale_offset, c='r', label='Harmonic Structure with Offset')
ax.set_xlabel('Logarithmic p-adic Number (base p)')
ax.set_ylabel('Note Names')
ax.set_title('Logarithmic p-adic Numbers - Harmonic Structure')
ax.legend()

# Get the maximum pitch class and octave from the harmonic structure
max_pitch_class = max(harmonic_structure)
max_octave = max([pitch_class // 12 for pitch_class in harmonic_structure])

# Set custom labels for y-axis to display note names with octave
ax.set_yticks(list(note_mapping.keys()))
ax.set_yticklabels(generate_notes(list(note_mapping.keys()), note_mapping))

plt.show
```
